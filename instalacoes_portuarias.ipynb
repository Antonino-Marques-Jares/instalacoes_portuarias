# Descobrindo latitude e longitude dos Municípios com Instlações Portuárias

```python

# origem dados dos estados e municipios https://github.com/kelvins/municipios-brasileiros/blob/main/csv/municipios.csv
df_estado = pd.read_csv('estados.csv', sep=',')
df_municipios = pd.read_csv('municipios.csv', sep=',')
```

```python

# adicionando UF com um Join entre Df_Estado e Df_Municipio
df_municipios = df_municipios.join(
    df_estado.set_index('codigo_uf')[['uf']],
    on='codigo_uf',
    how='left'
)
df_municipios.head()
```

```python
# Renomear a coluna 'nome' para 'municipio'
df_municipios = df_municipios.rename(columns={'nome': 'municipio'})
# Atribuindo o reordenamento ao DataFrame
nova_ordem = ['uf','municipio', 'latitude','longitude']
df_municipios = df_municipios[nova_ordem]
# Renomear as colunas
df_municipios = df_municipios.rename(columns={'uf': 'UF'})
df_municipios = df_municipios.rename(columns={'municipio': 'MUNICIPIO'})
df_municipios = df_municipios.rename(columns={'latitude': 'LATITUDE'})
df_municipios = df_municipios.rename(columns={'longitude': 'LONGITUDE'})
df_municipios.head(3)
```

# Municípios com Instalações Portuárias - ANTAQ
Link - https://dados.gov.br/dados/conjuntos-dados/instalacoes-portuarias

```python
# Primeiro, vamos criar um dicionário de mapeamento entre nomes de estados e siglas
estados_brasil = {
    'Acre': 'AC',
    'Alagoas': 'AL',
    'Amapá': 'AP',
    'Amazonas': 'AM',
    'Bahia': 'BA',
    'Ceará': 'CE',
    'Distrito Federal': 'DF',
    'Espírito Santo': 'ES',
    'Goiás': 'GO',
    'Maranhão': 'MA',
    'Mato Grosso': 'MT',
    'Mato Grosso do Sul': 'MS',
    'Minas Gerais': 'MG',
    'Pará': 'PA',
    'Paraíba': 'PB',
    'Paraná': 'PR',
    'Pernambuco': 'PE',
    'Piauí': 'PI',
    'Rio de Janeiro': 'RJ',
    'Rio Grande do Norte': 'RN',
    'Rio Grande do Sul': 'RS',
    'Rondônia': 'RO',
    'Roraima': 'RR',
    'Santa Catarina': 'SC',
    'São Paulo': 'SP',
    'Sergipe': 'SE',
    'Tocantins': 'TO'
}

# Adiciona uma coluna com a sigla UF no df_portos
df_portos['UF_SIGLA'] = df_portos['NOUF'].map(estados_brasil)

# Agora fazemos o merge (join) entre os dataframes
df_final = pd.merge(
    df_municipios,
    df_portos,
    how='inner',
    left_on=['UF', 'MUNICIPIO'],
    right_on=['UF_SIGLA', 'NOMunicipio']
)

# Seleciona apenas as colunas desejadas
df_final = df_final[[
    'UF', 
    'MUNICIPIO', 
    'LATITUDE', 
    'LONGITUDE',
    'NomeInstPortuaria',
    'Tipo'
]]

# Renomeia as colunas para o padrão desejado
df_final = df_final.rename(columns={
    'MUNICIPIO': 'MUNICIPIO',
    'NomeInstPortuaria': 'NOME_INSTALACAO_PORTUARIA',
    'Tipo': 'TIPO_INSTALACAO'
})

# Se precisar adicionar o endereço completo (opcional)
df_final['ENDERECO'] = df_final.apply(
    lambda row: f"{row['NOME_INSTALACAO_PORTUARIA']}, {row['MUNICIPIO']}, {row['UF']}",
    axis=1
)

# Renomeando as colunas específicas
df_final = df_final.rename(columns={
    'NOME_INSTALACAO_PORTUARIA': 'INSTALACAO_PORTUARIA',
    'TIPO_INSTALACAO': 'TIPO'
})

# Mostra o resultado
print(df_final.head())
```

```python
import json

# Converter o DataFrame para GeoJSON
features = []
for _, row in df_final.iterrows():
    feature = {
        "type": "Feature",
        "properties": {
            "instalacao": row['INSTALACAO_PORTUARIA'],
            "tipo": row['TIPO'],
            "municipio": row['MUNICIPIO'],
            "estado": row['UF'],
            "endereco": f"{row['INSTALACAO_PORTUARIA']}, {row['MUNICIPIO']}, {row['UF']}"
        },
        "geometry": {
            "type": "Point",
            "coordinates": [row['LONGITUDE'], row['LATITUDE']]
        }
    }
    features.append(feature)

geo_json = {
    "type": "FeatureCollection",
    "features": features
}

# Converter o GeoJSON para string JSON
geo_json_str = json.dumps(geo_json, ensure_ascii=False)

html_template = f"""
<!DOCTYPE html>
<html>
<head>
    <title>Mapa de Instalações Portuárias</title>
    <meta charset="utf-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    
    <!-- Leaflet CSS -->
    <link rel="stylesheet" href="https://unpkg.com/leaflet@1.7.1/dist/leaflet.css" />
    <link rel="stylesheet" href="https://unpkg.com/leaflet.markercluster@1.4.1/dist/MarkerCluster.css" />
    <link rel="stylesheet" href="https://unpkg.com/leaflet.markercluster@1.4.1/dist/MarkerCluster.Default.css" />
    
    <style>
        body {{
            margin: 0;
            padding: 0;
        }}
        #map {{
            width: 100%;
            height: 100vh;
        }}
        .popup-content {{
            font-family: Arial, sans-serif;
            max-width: 300px;
        }}
        .popup-title {{
            font-weight: bold;
            margin-bottom: 5px;
            color: #2c3e50;
        }}
        .popup-field {{
            margin: 3px 0;
        }}
    </style>
</head>
<body>
    <div id="map"></div>

    <!-- Leaflet JS -->
    <script src="https://unpkg.com/leaflet@1.7.1/dist/leaflet.js"></script>
    <script src="https://unpkg.com/leaflet.markercluster@1.4.1/dist/leaflet.markercluster.js"></script>
    
    <script>
        // Inicializar o mapa
        var map = L.map('map').setView([-15.788, -47.879], 4);
        
        // Adicionar base layer (OpenStreetMap)
        L.tileLayer('https://{{s}}.tile.openstreetmap.org/{{z}}/{{x}}/{{y}}.png', {{
            attribution: '&copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors'
        }}).addTo(map);
        
        // Dados dos portos
        var portosData = {geo_json_str};
        
        // Configurar clusters
        var clusters = L.markerClusterGroup();
        
        // Adicionar marcadores ao cluster
        L.geoJSON(portosData, {{
            pointToLayer: function(feature, latlng) {{
                return L.marker(latlng);
            }},
            onEachFeature: function(feature, layer) {{
                var popupContent = `
                    <div class="popup-content">
                        <div class="popup-title">${{feature.properties.instalacao}}</div>
                        <div class="popup-field"><b>Tipo:</b> ${{feature.properties.tipo}}</div>
                        <div class="popup-field"><b>Município:</b> ${{feature.properties.municipio}}</div>
                        <div class="popup-field"><b>Estado:</b> ${{feature.properties.estado}}</div>
                        <div class="popup-field"><b>Endereço:</b> ${{feature.properties.endereco}}</div>
                    </div>
                `;
                layer.bindPopup(popupContent);
            }}
        }}).addTo(clusters);
        
        // Adicionar clusters ao mapa
        map.addLayer(clusters);
        
        // Ajustar o zoom
        map.fitBounds(clusters.getBounds());
    </script>
</body>
</html>
"""

# Salvar o arquivo HTML
with open('mapa_portos.html', 'w', encoding='utf-8') as f:
    f.write(html_template)
```
